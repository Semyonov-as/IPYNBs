```python
from matplotlib import pylab as plt
from scipy import linalg
import numpy as np

%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
A1 = np.array([[1, 1], [1, 15]])
b1 = np.array([f(1), f(15)])
```

```python
w1 = linalg.solve(A1, b1)
```

```python
def g1(x):
    return w1[0] + w1[1]*x
```

```python
x = np.arange(1, 15, 0.5)
```

```python
plt.plot(x, f(x), x, g1(x))
```

```python
A2 = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b2 = np.array([f(1), f(8), f(15)])

w2 = linalg.solve(A2, b2)
```

```python
def g2(x):
    return w2[0] + w2[1]*x + w2[2]*x*x
```

```python
plt.plot(x, f(x), x, g2(x))
```

```python
A3 = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b3 = np.array([f(1), f(4), f(10), f(15)])

w3 = linalg.solve(A3, b3)
```

```python
def g3(x):
    return w3[0] + w3[1]*x + w3[2]*x*x + w3[3]*x*x*x
```

```python
plt.plot(x, f(x), x, g3(x))
```
