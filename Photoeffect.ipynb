```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (12, 12))
ax[0][0].scatter(x=l1['Vf'], y=l1['V'], marker='+', c = 'r', label='Эксперимент')
ax[0][0].set_xlim(xmax=l1['Vf'][0] + 0.003, xmin=float(l1['Vf'][-1:])-0.003)
ax[0][0].set_ylim(ymax=l1['V'][0] + 0.001, ymin=float(l1['V'][-1:])-0.001)
ax[0][0].set_title('$\lambda = 6402 \; \AA$', fontsize= 20)
fp1, _, _, _, _ = sp.polyfit(l1['Vf'], l1['V'], 3, full = True)
x1 = np.linspace(float(l1['Vf'][-1:]), l1['Vf'][0], 100)
f1 = sp.poly1d(fp1)
ax[0][0].plot(x1, f1(x1), 'b--', label='Приближение')
ax[0][0].legend()
ax[0][0].grid(linestyle="--", linewidth=0.5, color='y', zorder=-10)
ax[0][0].set_xlabel('$V_{ф}, В$', fontsize=14)
ax[0][0].set_ylabel('$\sqrt{I}\sim V, В$', fontsize=14)
ax[0][0].yaxis.set_minor_locator(AutoMinorLocator(4))
ax[0][0].xaxis.set_minor_locator(AutoMinorLocator(5))
ax[0][0].xaxis.set_major_locator(MultipleLocator(0.005))
ax[0][0].yaxis.set_major_locator(MultipleLocator(0.002))

ax[0][1].scatter(x=l2['Vf'], y=l2['V'], marker='+', c = 'r', label='Эксперимент')
ax[0][1].set_xlim(xmax=l2['Vf'][0] + 0.1, xmin=float(l2['Vf'][-1:])-0.1)
ax[0][1].set_ylim(ymax=l2['V'][0] + 0.01, ymin=0)
ax[0][1].set_title('$\lambda = 5852 \; \AA$', fontsize= 20)
fp2, _, _, _, _ = sp.polyfit(l2['Vf'], l2['V'], 3, full = True)
x2 = np.linspace(float(l2['Vf'][-1:]), l2['Vf'][0], 100)
f2 = sp.poly1d(fp2)
ax[0][1].plot(x2, f2(x2), 'b--', label='Приближение')
ax[0][1].legend()
ax[0][1].grid(linestyle="--", linewidth=0.5, color='y', zorder=-10)
ax[0][1].set_xlabel('$V_{ф}, В$', fontsize=14)
ax[0][1].set_ylabel('$\sqrt{I}\sim V, В$', fontsize=14)
ax[0][1].xaxis.set_major_locator(MultipleLocator(0.15))
ax[0][1].xaxis.set_minor_locator(AutoMinorLocator(3))
ax[0][1].yaxis.set_major_locator(MultipleLocator(0.025))
ax[0][1].yaxis.set_minor_locator(AutoMinorLocator(5))

ax[1][0].scatter(x=l3['Vf'], y=l3['V'], marker='+', c = 'r', label='Эксперимент')
ax[1][0].set_xlim(xmax=l3['Vf'][0] + 0.05, xmin=float(l3['Vf'][-1:])-0.05)
ax[1][0].set_ylim(ymax=l3['V'][0] + 0.01, ymin=0-0.01)
ax[1][0].set_title('$\lambda = 5400 \; \AA$', fontsize= 20)
fp3, _, _, _, _ = sp.polyfit(l3['Vf'], l3['V'], 3, full = True)
x3 = np.linspace(float(l3['Vf'][-1:]), l3['Vf'][0], 100)
f3 = sp.poly1d(fp3)
ax[1][0].plot(x3, f3(x3), 'b--', label='Приближение')
ax[1][0].legend()
ax[1][0].grid(linestyle="--", linewidth=0.5, color='y', zorder=-10)
ax[1][0].set_xlabel('$V_{ф}, В$', fontsize=14)
ax[1][0].set_ylabel('$\sqrt{I}\sim V, В$', fontsize=14)
ax[1][0].xaxis.set_major_locator(MultipleLocator(0.05))
ax[1][0].xaxis.set_minor_locator(AutoMinorLocator(5))
ax[1][0].yaxis.set_major_locator(MultipleLocator(0.025))
ax[1][0].yaxis.set_minor_locator(AutoMinorLocator(5))

ax[1][1].scatter(x=l4['Vf'], y=l4['V'], marker='+', c = 'r', label='Эксперимент')
ax[1][1].set_xlim(xmax=l4['Vf'][0] + 0.05, xmin=float(l4['Vf'][-1:])-0.05)
ax[1][1].set_ylim(ymax=l4['V'][0] + 0.01, ymin=0-0.01)
ax[1][1].set_title('$\lambda = 4916 \; \AA$', fontsize= 20)
fp4, _, _, _, _ = sp.polyfit(l4['Vf'], l4['V'], 3, full = True)
x4 = np.linspace(float(l4['Vf'][-1:]), l4['Vf'][0], 100)
f4 = sp.poly1d(fp4)
ax[1][1].plot(x4, f4(x4), 'b--', label='Приближение')
ax[1][1].legend()
ax[1][1].grid(linestyle="--", linewidth=0.5, color='y', zorder=-10)
ax[1][1].set_xlabel('$V_{ф}, В$', fontsize=14)
ax[1][1].set_ylabel('$\sqrt{I}\sim V, В$', fontsize=14)
ax[1][1].xaxis.set_major_locator(MultipleLocator(0.05))
ax[1][1].xaxis.set_minor_locator(AutoMinorLocator(5))
ax[1][1].yaxis.set_major_locator(MultipleLocator(0.025))
ax[1][1].yaxis.set_minor_locator(AutoMinorLocator(5))

def circle(x, y, radius, ax):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text, ax):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='black', fontsize=14)


#approx
circle(-0.68, 0, 0.01, ax[1][1])
text(-0.68, 0.025, "$V_0 = -0.68$", ax[1][1])
circle(-0.637, 0, 0.01, ax[1][0])
text(-0.637, 0.025, "$V_0 = -0.637$", ax[1][0])
circle(-0.318, 0, 0.001, ax[0][0])
text(-0.318, 0.0015, "$V_0 = -0.318$", ax[0][0])
fig.tight_layout()
fig.savefig("graphs")
```

```python
W = [2*np.pi*3*10**8/x for x in [L1, L3, L4]]
V = [0.318, 0.637, 0.68]
fpV, _, _, _, _ = sp.polyfit(W, V, 1, full = True)
xW = [W[0]-1*10**14, W[1], W[2]+ 1*10**14, 100]
xV = np.linspace(W[0]-1*10**14, W[2]+ 1*10**14, 100)
fV = sp.poly1d(fpV)

fi, ax = plt.subplots(figsize = (10, 6))


ax.grid(linestyle="--", linewidth=0.5, color='y', zorder=-10)
ax.set_xlabel(r'$\omega, 10^{15} \; \frac{рад}{с}$', fontsize=18)
ax.set_ylabel('$V_0, В$', fontsize=18)
ax.yaxis.set_minor_locator(AutoMinorLocator(5))
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.xaxis.set_major_locator(MultipleLocator(0.1*10**15))
ax.yaxis.set_major_locator(MultipleLocator(0.1))
ax.scatter(x=W, y=V, marker='o', c = 'y', label='Эксперимент')
ax.set_xlim(xmax=W[2] + 1*10**14, xmin=W[0]-1*10**14)
ax.set_ylim(ymax=V[2] + 0.1, ymin=V[0]-0.1)
ax.set_title(r'$V_0 = \frac{h\omega - A_{вых}}{e}$', fontsize= 20)
ax.plot(xW, fV(xW), 'b', label='Приближение')
ax.legend()

H = fpV[0]*(1.6*10**(-18))
fi.tight_layout()
fi.savefig("V0")
A = -fpV[1]
H, A
```
